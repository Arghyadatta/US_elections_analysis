```python
# Lets analyze the dataset

# importing  packages

import plotly.plotly as py
from plotly.offline import plot
import pandas as pd
import plotly.graph_objs as go

#Reading the dataframe into df
df = pd.read_csv("data/elections.csv", sep= ';')

# This groups by Democrats and Republican votes for 2008 and 2012
data1 = df.groupby('State')['Democrats 08 (Votes)'].sum().to_frame().rename(columns = {'Democrats 08 (Votes)':'Votes'})
data2 = df.groupby('State')['Democrats 12 (Votes)'].sum().to_frame().rename(columns = {'Democrats 12 (Votes)':'Votes'})
data3 = df.groupby('State')['Republicans 08 (Votes)'].sum().to_frame().rename(columns = {'Republicans 08 (Votes)':'Votes'})
data4 = df.groupby('State')['Republicans 12 (Votes)'].sum().to_frame().rename(columns = {'Republicans 12 (Votes)':'Votes'})

#-------------------------Vote Distributions for 2008 US elections--------------------------------
trace1 = go.Bar(
            x= data1.index.tolist(),
            y= data1.Votes.tolist(),
            name = 'Democratic Party'
    )

trace2 = go.Bar(
            x= data3.index.tolist(),
            y= data3.Votes.tolist(),
            name = 'Republican Party'
    )


data = [trace1, trace2]
layout = go.Layout(
    barmode='stack'
)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='stacked-bar')
    
    
```

```python

#---------------Vote Distributions for 2012 US elections--------------
trace1 = go.Bar(
            x= data2.index.tolist(),
            y= data2.Votes.tolist(),
            name = 'Democratic Party'
    )

trace2 = go.Bar(
            x= data4.index.tolist(),
            y= data4.Votes.tolist(),
            name = 'Republican Party'
    )


data = [trace1, trace2]
layout = go.Layout(
    barmode='stack'
)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='stacked-bar')    
```

```python
# Distributions of population and median earnings across counties in US


import plotly.plotly as py
import pandas as pd

import numpy as np

columns = ['Median Earnings 2010', 'White', 'Black', 'Hispanic', 'Asian', 'Amerindian', 'Other', 'White_Asian']

df['White'] = df['White']*df['Total Population']/100
df['Black'] = df['Black']*df['Total Population']/100
df['Hispanic'] = df['Hispanic']*df['Total Population']/100
df['Asian'] = df['Asian']*df['Total Population']/100
df['Amerindian'] = df['Amerindian']*df['Total Population']/100
df['Other'] = df['Other']*df['Total Population']/100
df['White_Asian'] = df['White_Asian']*df['Total Population']/100

#data1 = df.groupby(['ST', 'County'])['Median Earnings 2010'].median().apply(np.round_)
#data2 = df.groupby(['ST','County'])['White'].mean().apply(np.ceil)
#data3 = df.groupby(['ST','County'])['Black'].mean().apply(np.ceil)
#data4 = df.groupby(['ST','County'])['Hispanic'].mean().apply(np.ceil)
#data5 = df.groupby(['ST','County'])['Asian'].mean().apply(np.ceil)
#data6 = df.groupby(['ST','County'])['Amerindian'].mean().apply(np.ceil)
#data7 = df.groupby(['ST','County'])['Other'].mean().apply(np.ceil)
#data8 = df.groupby(['ST','County'])['White_Asian'].mean().apply(np.ceil)

data1 = df.groupby('ST')['Median Earnings 2010'].median().apply(np.round_)
data2 = df.groupby('ST')['White'].mean().apply(np.ceil)
data3 = df.groupby('ST')['Black'].mean().apply(np.ceil)
data4 = df.groupby('ST')['Hispanic'].mean().apply(np.ceil)
data5 = df.groupby('ST')['Asian'].mean().apply(np.ceil)
data6 = df.groupby('ST')['Amerindian'].mean().apply(np.ceil)
data7 = df.groupby('ST')['Other'].mean().apply(np.ceil)
data8 = df.groupby('ST')['White_Asian'].mean().apply(np.ceil)

data = pd.concat([data1,data2,data3,data4,data5,data6,data7,data8], axis=1)

data
```

```python
import plotly.plotly as py
import pandas as pd

data=data.reset_index()

for col in data.columns:
    data[col] = data[col].astype(str)

scl = [[0.0, 'rgb(245,247,250)'],[0.2, 'rgb(230,230,240)'],[0.4, 'rgb(207,212,227)'],\
            [0.6, 'rgb(175,180,220)'],[0.8, 'rgb(130,137,190)'],[1.0, 'rgb(84,39,143)']]

data['text'] = data['ST'] + '<br>' +\
'Median Salary: '+data['Median Earnings 2010']+' White '+data['White']+'<br>'+\
'Black: '+data['Black']+' Hispanic: ' + data['Hispanic']+'<br>'+\
' Asian: '+data['Asian']+' Amerindian: '+data['Amerindian']  

#data = data.drop(labels=['level_0','index'],axis=1)
dt = [ dict(
        type='choropleth',
        colorscale = scl,
        autocolorscale = False,
        locations = data['ST'],
        z = data['Median Earnings 2010'].astype(float),
        locationmode = 'USA-states',
        text = data['text'],
        marker = dict(
            line = dict (
                color = 'rgb(255,255,255)',
                width = 2
            ) ),
        colorbar = dict(
            title = "USD")
        ) ]

layout = dict(
        title = '2010 US Median Earnings by State<br>(Hover for breakdown)',
        geo = dict(
            scope='usa',
            projection=dict( type='albers usa' ),
            showlakes = True,
            lakecolor = 'rgb(255, 255, 255)'),
             )
    
fig = dict( data=dt, layout=layout )
py.iplot( fig, filename='d3-cloropleth-map' )
```

```python

df['Violent.crime']= df['Violent.crime'] * df['Total Population']/100000

scl = [[0.0, 'rgb(245,247,250)'],[0.2, 'rgb(230,230,240)'],[0.4, 'rgb(207,212,227)'],\
            [0.6, 'rgb(175,180,220)'],[0.8, 'rgb(130,137,190)'],[1.0, 'rgb(84,39,143)']]

data1 = df.groupby('ST')['Violent.crime'].mean().apply(np.round_).to_frame()
data1['crime_norm'] = (data1['Violent.crime'] - data1['Violent.crime'].median())/(data1['Violent.crime'].max() - data1['Violent.crime'].min())

data = pd.concat([data1,data2,data3,data4,data5,data6,data7,data8], axis=1)

data.reset_index(inplace=True)

for col in data.columns:
    data[col] = data[col].astype(str)
    
data['text'] = data['ST'] + '<br>' +\
'Violent Crime: '+data['Violent.crime']+' White '+data['White']+'<br>'+\
'Black: '+data['Black']+' Hispanic: ' + data['Hispanic']+'<br>'+\
' Asian: '+data['Asian']+' Amerindian: '+data['Amerindian']  

#data = data.drop(labels=['level_0','index'],axis=1)
dt = [ dict(
        type='choropleth',
        colorscale=scl,
        autocolorscale = False,
        locations = data['ST'],
        z = data['crime_norm'].astype(float),
        locationmode = 'USA-states',
        text = data['text'],
        marker = dict(
            line = dict (
                color = 'rgb(255,255,255)',
                width = 2
            ) ),
        colorbar = dict(
            title = "No. of Crimes")
        ) ]

layout = dict(
        title = '2010 US Violent Crimes by State<br>(Hover for breakdown)',
        geo = dict(
            scope='usa',
            projection=dict( type='albers usa' ),
            showlakes = True,
            lakecolor = 'rgb(255, 255, 255)'),
             )
    
fig = dict( data=dt, layout=layout )
py.iplot( fig, filename='d3-cloropleth-map' )
```
